  > \>>> c = c + 5

  > \>>> print(c)   
    > [Count (step: 1)] 17   
    > \>>> print(d)

- 요구사항 6. 다음과 같은 관계연산 (>,<,==)이 수행될 수 있도록 Counter 클래스 내에 관련 메소드를 추가하시오.  
    > \>>> print(c) > 10   
    > True    
    > \>>> print(d) > 10   
    > False   
    > \>>> print(c) < 10   
    > False   
    > \>>> print(d) < 10   
    > True   
    > \>>> print(c) == 17   
    > True   
    > \>>> print(d) != 9   
    > False    
    
  - 해당 문제는 print()가 NoneType이고 클래스에서 호출이 불가능하기 때문에   
    print(c) > 10 을 print(c>10)으로 고려하고 처리했다.

```python
class Counter:
    def __init__(self,count,step=1):
        self.count=count
        self.step=step
    
    def __str__(self):
        return str('[Count (step: ' + str(self.step) + ')] ' + str(self.count))
    
    def incr(self):
        self.count = self.count + self.step
    
    def __call__(self):
        self.incr()
        
    def __add__(self,B):
        self.count = self.count + B
        return self
    
    def __sub__(self,B):
        self.count = self.count - B
        return self
    
    def __gt__(self,B):
        return self.count > B
    
    def __lt__(self,B):
        return self.count < B
    
    def __eq__(self,B):
        return self.count == B
    
    def __ne__(self, B):
        return self.count != B
        
if __name__ == "__main__" :
    c = Counter(10)
    d = Counter(10,2)
    print('# 3-1,2')
    print(c)
    print(d)
    print()
    
    print('# 3-3')
    c.incr()
    d.incr()
    print(c)
    print(d)
    print()
    
    print('# 3-4')
    c()
    d()
    print(c)
    print(d)
    print()
    
    print('# 3-5')
    c = c + 5
    d = d - 5
    print(c)
    print(d)
    print()
    
    print('# 3-6')
    print(c > 10 )
    print(d > 10)
    print(c < 10)
    print(d < 10)
    print(c == 17)
    print(d != 9)
```

#### [문제4]  
다음은 내장 자료형 list를 서브클래싱하여 만든 MySet 클래스 정의 내용이다. 다음 클래스 정의에서 \__init__(), \__str()__(), elimicate_duplicate()의 세 개의 메소드 코드 내용을 자신이 다른 사람에게 가르친다고 생각하며 설명해보시오.  
   - MySet은 집합(Set) 자료형을 정의하려는 의도하에 만들어진 클래스이다.

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
```

#### [설명]  
- 1) \__init__()  
    - 생성자가 호출될 때 실행되는 함수이다. 해당 MySet 클래스는 list의 서브클래스로 list에 존재하는 append를 사용가능하다. 이 때 append는 list에 append의 인자로 들어온 값을 list에 추가한다. 그래서 ' for e in l: '에서 l에서 원소를 하나씩 e로 넣어 이를 list의 서브 클래스인 MySet클래스의 객체 인스턴스에 하나씩 저장한다. 후에 클래스 이름공간에 존재하는 정적메소드 eliminate_dupliacte를 클래스 이름(MySet)을 이용해서 호출하므로써 객체 인스턴스에 중복된 원소를 제거해준다.   
- 2) \__str__()  
    - 객체 인스턴스를 print()를 이용해서 출력할 때 print()의 괄호 안이 str()과정을 거치면서 \__str__함수가 호출된다. 위의 코드는 객체 인스턴스 내의 원소를 특정 형태로 출력하고 싶어서 우선 result에 처음 클래스 이름를 표기하기 위한 'MySet: '을 저장하고 그 뒤에 for문을 써서 객체 인스턴스의 요소 하나씩 ','와 함께 result에 연속적으로 이어지게 저장시켰다. 그리고 마지막 원소가 출력될 때 같이 출력된 콤마(,)를 지우기 위해 \[0:len(result)-2]까지 슬라이싱을 했고 뒤에 '}'를 result에 저장했다. 그리고 이 result를 반환시켰다. 
- 3) eliminate_duplicate()  
    - 클래스 이름공간에 존재하는 정적메소드이다. 그래서 다른 곳에서 호출할 때 클래스 이름을 이용해서 호출한다.( + python3에서 self를 통해서도 호출가능하게 설정되어 있다.)  
    이 함수는 인자로 리스트를 받아서 빈 리스트 s에  
      
          for e in l:     
            if e not in s: 
                  s.append(e)  
                  
    위와 같은 과정을 거치면서 l에서 리스트 요소값을 하나씩 e에 넣어서 e와 같은 값이 s에 존재하지않으면 s에 추가하면서 l에서 중복된 원소들을 빼고 s에 저장한다.   
    그리고 l을 슬라이싱을 통해 다시 비우고 s의 값들을 l에 집어넣어서 최종적으로 중복된 원소가 없는 리스트를 만들어준다.
  

#### [문제5]   
4번 문제에 정의된 MySet 클래스에 메소드를 추가하여 다음 각 요구사항 모두를 만족시키는 코딩을 제시하시오  
- 정답을 각 요구사항별로 입력할 필요 없이 요구사항 3개 전체에 대해 1개의 MySet 클래스 정의 코드를 제시하고 주석등으로 해당 코드를 설명한다.  
- 요구사항 1. | 연산으로 두 집합의 합집합을 반환한다.  
    > \>>> u = s | t   
    > \>>> print(u)  
    > MySet: {1, 2, 3, 4, 5, 6, 7, 8, 9}  
- 요구사항 2. & 연산으로 두 집합의 교집합을 반환한다.  
    > \>>> u = s & t  
    > \>>> print(u)  
    > MySet: {2, 3}  
- 요구사항 3. - 연산으로 두 집합의 차집합을 반환한다. 
    > \>>> s = MySet([1, 2, 3])   
    > \>>> t = MySet([3, 4, 5])  
    > \>>> u = s - t  
    > \>>> print(u)   
    > MySet: {1, 2} 

```python
class MySet(list):
    def __init__(self, l):
        for e in l:
            self.append(e)
        MySet.eliminate_duplicate(self)
    
    def __str__(self):
        result = "MySet: {"
        for e in self:
            result = result + str(e) + " ,"
        result = result[0:len(result)-2] + "}"
        return result

    def __or__(self,B):
        s=[]
        for e in self:
            s.append(e)
        for e in B:
            if e not in s:
                s.append(e)
        s.sort()
        a = MySet(s)
        return a
    
    def __and__(self,B):
        s=[]
        for e in self:
            for k in B:
                if e==k :
                    s.append(e)
        s.sort()
        a = MySet(s)
        return a
    
    def __sub__(self,B):
        s=[]
        for e in self:
            if e not in B :
                s.append(e)
        s.sort()
        a = MySet(s)
        return a
    
    @staticmethod    
    def eliminate_duplicate(l):
        s = []
        for e in l:
            if e not in s:
                s.append(e)
        l[:] = []
        for e in s:
            l.append(e)

    
if __name__ == "__main__":
    s = MySet([1, 2, 2, 3])
    print(s)
    t = MySet([2, 3, 4, 5, 6, 7, 8, 8, 8, 8, 8, 9])
    print(t)
    print()
    
    print('# 5 - or')
    u = s | t
    print(u)
    print()
    
    print('# 5 - and')
    u = s & t
    print(u)
    print()
    
    print('# 5 - sub')
    s = MySet([1, 2, 3])
    t = MySet([3, 4, 5])
    u = s - t
    print(u)
```

#### [설명]  
- (1->or)  
    - 빈 리스트 s에 우선 self의 리스트 요소를 넣게 하고 후에 다시 반복문을 통해 B의 리스트 요소 중 s에 없는 값은 s에 append를 통해 추가하게 했다. 그리고 s.sort()를 통해 리스트 s를 정렬시켰다. 반환할 때 다시 MySet클래스의 객체 인스턴스로 반환해야하므로 a=MySet(s)를 통해 객체 인스턴스 a가 리스트 s를 인자로 가지도록 하고 a를 반환했다.  
- (2->and)  
    - 빈 리스트 s에 self의 리스트 요소들 중 B 리스트에 존재하면 s에 append를 통해 추가하게 했다. 그리고 s.sort()를 통해 s를 정렬시켰고 이를 MySet생성자의 인자로 넣어 그렇게 만들어진 객체 인스턴스를 반환하게 했다.  
- (3->sub)  
    - 빈 리스트 s에 self의 리스트 요소가 B 리스트에 존재하지 않으면 s에 append를 통해 추가해 차집합을 구현했다. 그리고 s.sort()를 통해 s를 정렬시켰고 이를 MySet생성자의 인자로 넣어 그렇게 만들어진 객체 인스턴스를 반환하게 했다.

     - 최다 출현 단어의 빈도수가 동일한 경우 모두 출력해주어야 함

 - 만약 등록된 URL이 없다면 getMaxFreqencyWords()의 반환 값은 None이 되어야 함.

    - 파라미터로 받은 단어와 유사도가 동일한 URL이 여러 개이면 해당 URL을 모두 출력하시오.  

```python
import requests
import string

class SearchEngine:
    
    def __init__(self,*args):
        self.urllist = list(args)
        self.Dicts=[]
    
    def addUrl(self,url):
        self.urllist.append(url)
    
    def removeUrl(self,url):
        if url in self.urllist:
            del self.urllist[self.urllist.index(url)]
    
    def listUrls(self):
        for url in self.urllist:
            print(url)
    
    def getWordsFrequency(self):
        self.Dicts=[]
        srclist=[]
        if self.urllist==[]:
            print({})
            return {}
        for i in range(len(self.urllist)):
            srclist.append(requests.get(self.urllist[i]).text)
        for i in range(len(self.urllist)):
            P = self.list_string_pure(self.text_parsing(srclist[i])) # P는 구두점 제거, 비어있는 리스트 원소값 제거한 리스트
            D = self.dict_countWords(P)
            self.Dicts.append(D)
            
        return self.Dicts
    
    def getMaxFrequencyWords(self):
        self.getWordsFrequency()
        MaxWords = []
        
        if self.urllist == []:
            return None
        print()
        for i in range(len(self.urllist)):
            MaxWords_perUrl = []
            Result=[]
            list1 = list(self.Dicts[i].keys())
            list2 = list(self.Dicts[i].values())
            TupleD = tuple(zip(list1,list2))
            list2.sort(reverse=True) # 키값이 들어가 있는 list2를 역순으로 정렬
            
            # 정렬된 list2와 원래 사전 객체 순서로 저장된 튜플 TupleD[k][1]을 
            ## 비교했을 때 같으면 그 때의 key값을 넣으면 빈도수 가장 높은 
            ### 단어 순서로 Result 리스트에 추가된다.
            #### 따라서 많이 출현한 단어 3개를 뽑을 때 0부터 2까지의 Result 리스트 
            ##### 요소값을 뽑아내면 된다.
            for j in range(len(list2)):
                for k in range(len(TupleD)):
                    if list2[j]==TupleD[k][1]: 
                        Result.append(TupleD[k][0])
            print(self.urllist[i])
            for j in range(len(list2)-1):
                MaxWords_perUrl.append(Result[j])
                print(Result[j])
                if list2[j]!=list2[j+1]:
                    break;
            MaxWords.append(MaxWords_perUrl)
        return MaxWords
    
    def searchUrlByWord(self,word):
        self.getWordsFrequency()
        similarity=[0 for i in range(len(self.urllist))]
        s_list = word.split()
        s_list = self.list_string_pure(s_list)
        # 사전 객체에 같은 단어가 있는 경우 단어의 count값을 similarity에 계속 더하고
        ## similarity 가 가장 높게 나온 사이트를 가장 먼저 출력한다.
        for i in range(len(self.urllist)): 
            for j in s_list:
                for k in self.Dicts[i].keys():
                    if j == k:
                        similarity[i] += self.Dicts[i][j]
        list=[]
        # 유사도가 똑같은 url이 두번 출력 안되게 한 번 출력된 url을 key값으로 저장할 사전객체
        ## 해당 url에 해당되는 value 가 1이 되면 출력이 안되고 다음 url을 확인하게 함
        tempDicts={}
        for i in range(len(self.urllist)):
            tempDicts[self.urllist[i]]=0

        for i in range(len(self.urllist)):
            list.append(similarity[i])
        temp=[i for i in similarity]
        temp.sort(reverse=True)
       
        for i in range(len(temp)):
            for j in range(len(list)):
                if temp[i]==list[j]:
                    if tempDicts[self.urllist[j]]==0:
                        print("유사도:",list[j],'->' ,self.urllist[j])
                        tempDicts[self.urllist[j]]=1
                        if (i<len(list)-1) and (temp[i]!=temp[i+1]) :
                            return
                        break
                    elif tempDicts[self.urllist[j]]==1:
                        continue
    
    # 함수1 : JS,RSS,CSS 코드 제거 , Tag 제거
    def text_parsing(self,src):
        i=0
        Pure_txt=''
        while i< len(src): 
            if src[i]=='<': # ①
                if src[i:i+7]=='<script': #①-②-1
                    i+=7
                    while src[i:i+9]!='</script>':
                        i+=1
                    i+=9
                elif src[i:i+6]=='<style': #①-②-2
                    i+=6
                    while src[i:i+8]!='</style>':
                        i+=1
                    i+=8
                else :   #①-②-3
                    i+=1
                    while src[i]!='>':
                        i+=1
                    i+=1
            else:
                Pure_txt += src[i]
                i+=1
        return Pure_txt.split()
    
    #함수2 : 순수 텍스트가 요소로 저장되어 있는 리스트에서 단어의 출현빈도를 사전으로 반환
    def stopwords_filter(self,list):
        with open("E_stopwords.txt",'r') as f:
            S_E_stopwords=f.read()
        with open("Kor_stopwords.txt",'r') as f:
            S_Kor_stopwords=f.read()
        with open("sign.txt",'r') as f:
            S_sign_words=f.read()
        E_stopwords=S_E_stopwords.split()
        Kor_stopwords=S_Kor_stopwords.split('\n')
        sign_words=S_sign_words.split()
        # 영어 불용어 필터
        for i in range(len(list)):
            for j in range(len(E_stopwords)):
                if list[i].lower()==E_stopwords[j]:
                    list[i]=''
        # 한글 불용어 필터 
        ## 단어의 끝에서부터 불용어와 똑같은 길이만큼 슬라이싱해 불용어와 일치시 불용어 부분을 제거하여 저장
        for i in range(len(list)):
            for j in range(len(Kor_stopwords)):
                if len(list[i])>=len(Kor_stopwords[j]):
                    if list[i][-len(Kor_stopwords[j]):]==Kor_stopwords[j]:
                        list[i]=list[i][:-len(Kor_stopwords[j])]


        # 특수문자 필터
        for i in range(len(list)):
            for j in range(len(sign_words)):
                list[i]=list[i].replace(sign_words[j],'')

        return list
    
    #함수3 : 리스트 내 str 원소의 구두점 제거, 리스트 원소 값이 비어있으면 제거
    def list_string_pure(self,list):
        for i in range(len(list)):
            for j in range(len(string.punctuation)):
                list[i]=list[i].replace(string.punctuation[j],'')

        list = self.stopwords_filter(list) # 불용어 제거

        i=0
        while i!=len(list):
            if list[i]=='':
                del list[i]
                continue
            i+=1
        return list
    
    #함수4: 단어 count한 후 사전객체로 반환
    def dict_countWords(self,list):
        d={}
        for i in range(len(list)):
            d[list[i]]=0
        for i in range(len(list)):
            d[list[i]]=d[list[i]]+1
        return d
    
if __name__ == '__main__':
    w1 = SearchEngine('http://www.cnn.com', 'http://www.times.com', 'https://www.amazon.com')
    w2 = SearchEngine('http://www.cnn.com', 'http://www.times.com') 
    w3 = SearchEngine() 
    
    w1.addUrl('https://github.com')
    w3.addUrl('http://stackoverflow.com')
    
    w1.removeUrl('http://www.cnn.com')
    w2.removeUrl('http://stackoverflow.com')
    
    w2.listUrls()
    
    w1.getWordsFrequency() # 반환만 하게 하고 출력결과 확인이 용이하게
                           ## 밑에 따로 호출해서 Output보여주게 함.
    
    w1.getMaxFrequencyWords()
    print()
    w1.searchUrlByWord("characters")
```
