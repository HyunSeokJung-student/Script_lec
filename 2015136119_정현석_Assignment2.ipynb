- 1 or 2 or 3 or 4

```python
print(1 and 2 and 3 and 4)
print(1 or 2 or 3 or 4)
print(1 and 2 or 3 and 4)
print((1 and 2)or(3 and 4))
print(1 or 2 and 3 or 4)
print((1 or 2)and(3 or 4))
```

[설명]

<ul> and 연산자는 앞뒤의 피연산자가 둘 다 True인 것을 확인해야하므로 뒤의 피연산자를 반환한다.</ul>

이를 통해 위의 Expression을 해석하면 다음과 같다.  
<ul>1. 1 and 2 and 3 and 4 : and는 뒤의 피연산자까지 확인해야 True인지 False인지를 확정할 수 있기 때문에 최종 평가대상이 4이다.</ul>
<ul>2. 1 or 2 or 3 or 4 : or는 앞의 피연산자가 True이면 뒤 피연산자를 확인할 필요가 없으므로 최종 평가대상이 1이다.</ul>
<ul>3. 1 and 2 or 3 and 4 : 인터프리터가 문장을 읽을 때 왼쪽부터 읽어들이는데 1 and 2에서 2를 반환 후 2 or 3에서 2가 True이므로 뒤는 읽지 않고 최종 평가대상인 2를 반환한다.</ul>
<ul>4. (1 and 2) or (3 and 4) : 연산 우선순위에 의해 괄호 안의 연산부터 읽어 들인다. 1 and 2 는 뒤 피연산자인 2를 3 and 4 는 뒤 피연산자인 4를 반환한다. 그 후 2 or 4 연산을 하면 or 연산자에서 앞의 피연산자 2가 True이므로 최종 평가대상인 2가 출력된다.</ul>
<ul>5. 1 or 2 and 3 or 4 : 왼쪽부터 연산을 시작할 때 1 or 2에서 1이 True이므로 뒤의 대상인 and 3 or 4를 무시하고 최종 평가대상인 1을 출력한다.</ul>
<ul>6. (1 or 2) and (3 or 4) : 괄호 연산 내에서 or 연산자는 앞의 피연산자가 True이면 그 앞의 피연산자만 평가하므로  1 or 2는 1을, 3 or 4는 3을 반환한다. 괄호 연산이 끝난 후 1 and 3을 연산한 때 and 연산자는 뒤의 피연산자까지 True임을 확인해야하므로 최종 평가대상인 3을 반환한다.</ul>

### 문제 2
2) 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.
- 인자로 전달하는 문자열은 경로만 들어간다고 가정한다.
- 각 디렉토리와 파일을 구분하는 문자는 '/'로 가정한다.
- 반환하는 리스트의 첫번째 원소는 디렉토리이고 두번째 원소는 파일명이다.
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    - div_path('/usr/local/bin/python')  
        - ['/usr/local/bin','python']
    - div_path('/home/chulsoo/test.txt')  
        - ['/home/chulsoo','test.txt']
- [참고] 리스트(I) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 I.append(10)이다.

```python
def div_path(s):
    X = []
    Y = ''
    Z = []
    I=s.split('/') #1
    X=I.pop()      #2
    for i in range(0,len(I)):  #3
        if i==0:
            Y += str(I[i])
        else :
            Y += ('/'+str(I[i]))
    Z.append(Y)        #4
    Z.append(str(X))
    print(Z)
div_path('/usr/local/bin/python')
div_path('/home/chulsoo/test.txt')
```

[설명]  
디렉토리 경로명과 파일명을 구분하는 법은 /로 구분된 str 중 제일 마지막 요소만 빼서 반환하면 된다.  
<ul>1.그렇기 때문에 우선, split('/')함수를 통해 /로 구분된 str로 구성된 list I에 저장했다.  
2.이 리스트의 마지막 요소를 빼면 나머지 요소는 디렉토리 경로명 중에 '/'이 빠진 형태가 된다.  
    3.파일명인 리스트의 마지막 요소를 I.pop()을 통해 X에 저장한다.  
    4.split()함수를 거치는 와중에 사라진 '/'를 다시 넣어주기 위해 str형 Y에 Y+= ('/'+str(I[i]))를 통해 디렉토리 경로명을 str형으로 가지고 있는 변수 Y를 완성할 수 있다.  
  5.이렇게 X(파일명)와 Y(디렉토리명)으로 분리된 것을 새로운 리스트의 각요소 넣기 위해 Z.append()를 써서 디렉토리명, 파일명 순으로 요소를 추가한다.</ul>
  

### 문제 3
3) 두 개의 리스트를 인자로 받아서 그 두개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista,listb)를 작성하시오.  
- 인자로 전달하는 리스트 2 개에는 정수값만 들어간다고 가정하자.
- 함수 내에서 새로운 리스트를 만들어 그 리스트 내에 인자로 받은 두 리스트의 모든 원소를 넣어 반환한다.
- 반환하는 리스트에는 절대로 중복된 원소가 들어 있으면 안된다 (집합의 조건).
- 다음과 같은 실행 및 출력 결과가 도출되어야 한다.
    - list_union([1,2,3],[1,2,4])  
        - [1,2,3,4]
    - list_union([-10,-5,0,-1],[100,9,0,9])  
        - [-10,-5,-1,0,9,100]
    - list_union([0,1,2],[0,1,2])  
        - [0,1,2]
- [참고] 리스트(I) 내에 새로운 정수값 (예를 들어 10)을 넣는 방법은 I.append(10)이다.
- [참고] 임의의 정수값(x)이 리스트(I) 내에 존재하는지 판단하는 방법은 x in I이다.

```python
def list_union(lista,listb):
    L1=[] # 중복된 원소가 없는 A 집합
    L2=[] # 중복된 원소가 없는 B 집합
    L3=[] # A-B에 해당되는 원소를 저장할 리스트
    r=[] # 합집합을 저장할 리스트
    # ---------------------------------- 중복된 원소가 없는 리스트 L1 구하기 (by lista)
    for i in range(0,len(lista)-1):
        b=True
        for j in range(i+1,len(lista)):
            if lista[i]==lista[j]:
                b=False
                break
        if b==True:
            L1.append(lista[i])
    L1.append(lista[len(lista)-1]) # 마지막 인덱스 값은 바로 추가 (이전에 같은 수가 있었을 때 추가하지 않았기 때문에)
    # ---------------------------------- 중복된 원소가 없는 리스트 L2 구하기(by listb)
    for i in range(0,len(listb)-1):
        b=True
        for j in range(i+1,len(listb)):
            if listb[i]==listb[j]:
                b = False
                break
        if b==True:
            L2.append(listb[i])
    L2.append(listb[len(listb)-1])
    # -------------------------------- 리스트 L1의 요소 중 L2와 중복되지 않는 값들에 대한 리스트 L3
    for i in range(0,len(L1)):
            if (L1[i] in L2)== False:
                r.append(L1[i])
    
    r.extend(L2) # L3 + L2
    r.sort()     # 오름차순 정렬
    print(r)
    
list_union([1,2,3],[1,2,4])
list_union([-10,-5,0,-1],[100,9,0,9])
list_union([0,1,2],[0,1,2])
```

[설명]  
파라미터로 들어오는 두 개의 리스트 A와 B의 합집합을 구해야하는데 이를 위해서 다음의 과정을 거치면 된다.  
- 각각의 리스트에 중복되는 원소를 가지지 않도록 중복되는 원소를 걸러내 중복되는 원소가 없는 리스트(L1,L2)를 새로 만든다.  
- 중복되는 원소가 없는 리스트(L1,L2) 중 하나(L1)를 골라 그 리스트의 원소들이 남은 다른 리스트(L2)와 중복되는 원소가 있는지 확인한다. 중복되지 않는 원소들을 L3 리스트에 저장한다. (L3 : L1에서 L1과 L2의 교집합을 뺀 것)  
- L2와 L3을 더하고 정렬하면 최종적으로 A와 B의 합집합을 구할 수 있다.  

각 집합 A와 B를 중복되는 원소가 있어도 하나만 포함되게 하기 위해 A의 lista[i]와 인덱스가 i보다 큰 모든 리스트 값들을 비교해 중복되는 숫자가 없다면 b값을 True로 유지되게 해서 (있다면 b값 False) b가 True일 때 lista[i]를 새로운 리스트 L1의 요소로 append()통해 추가했다. L2도 마찬가지이다.  
  
이렇게 만들어진 L1과 L2의 합집합을 구하기 위해 L1의 원소 L1[i]가 L2에 존재하지 않으면(L1[i] in L2 : False) L3에 추가되도록 했다.

![image](https://github.com/JungHyunSuok/Script_lec/blob/master/10809.png?raw=true)

```python
s=input()
c='abcdefghijklmnopqrstuvwxyz'
for i in range(0,len(c)):
    if s.count(c[i])==0:
        print(-1,end=' ')
    else :
        print(s.find(c[i]),end=' ')
```

[설명]

QR Code "alphanumeric" 문자는 0123456789ABCDEFGHIJKLMNOPQRSTUVWXYZ\$%*+-./: 이다.  
    ![image](https://github.com/JungHyunSuok/Script_lec/blob/master/2675.png?raw=true)

```python
T=int(input())
for i in range(0,T) :
    R ,S= input().split()
    for j in range(0,len(S)):
        print(S[j]*int(R),end='')
    print()
```

```python
import requests
req = requests.get('http://cse.koreatech.ac.kr')
source = req.text #1
i=0
Pure_txt=''

while i<= (len(source)-1): #2
    if source[i]=='<':
        i+=1
        while source[i]!='>':
            i+=1
        i+=1
    else:
        Pure_txt += source[i]
        i+=1
P=Pure_txt.split()
print(len(P))
```
