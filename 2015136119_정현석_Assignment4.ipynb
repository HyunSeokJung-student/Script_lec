```python
s = '''pig ham
cat dog
ham bird
dog pig
'''
f = open("s.txt","w")
f.write(s);
f.close();
with open("s.txt",'r') as f:
    print('s.txt')
    print(f.read())
# 1-1 
list1=[]
with open("s.txt",'r') as f:
    list1=f.readlines()
    list1.sort()
with open("s1.txt",'w') as f:
    for l in list1: 
        f.writelines(l)
## 1-1 result
print("1-1 result")
with open("s1.txt",'r') as f:
    for line in f.readlines():
        print(line,end="")
# 1-2
list2=[]
L2=[]
i=0
with open("s.txt",'r') as f:
    line = f.readline()
    while line:
        L2 = line.split()
        L2.reverse()
        list2.append(' '.join(L2)+'\n')
        line = f.readline()
    list2.sort()
    for i in range(len(list2)):
        list2[i]=' '.join(reversed(list2[i].split())) + '\n'
with open("s2.txt",'w') as f:
    for l in list2: 
        f.writelines(l)
## 1-2 result
print("\n1-2 result")
with open("s2.txt",'r') as f:
    for line in f.readlines():
        print(line,end="")
        
# 1-3
list3=[]
s3=''
with open("s.txt",'r') as f:
    line = f.readline()
    while line:
        list3.append(line.split()[0])
        list3.append(line.split()[1])
        line = f.readline()
    for i in range(len(list3)):
        s3 = s3 + list3[i]
        if (i+1)%3==0 :
            s3 = s3 + '\n'
        else :
            if (i+1)!=len(list3):
                s3 = s3 + ' '
with open("s3.txt",'w') as f: 
    f.write(s3)
## 1-3 result
print("\n1-3 result")
with open("s3.txt",'r') as f:
    for line in f.readlines():
        print(line,end="")
```

#### 문제 3  
range() 함수와 유사한 frange() 함수를 다음 조건을 참고하여 만드시오.  

- 1) frange()의 인자 구성은 다음과 같이 range와 동일하지만 각 인수들은 음수를 받지 않는다고 가정한다.  
> range(stop)  
> range(start,stop[,step])  

- 2) frange() 함수의 인자에 대한 기본 시작(start) 값은 0.0이고, 기본 단계(step) 값은 0.1이다.  
- 3) frange 사용 예  
    - 3-1) frange(0.5)  
        - [0.0,0.1,0.2,0.3,0.4]  
    - 3-2) frange(1.0,2.0)  
        - [1.0,1.1,1.2,1.3,1.4,1.5,1.6,1.7,1.8,1.9]  
    - 3-3) frange(2.2,4.0,0.5)  
        - [2.2,2.7,3.2,3.7]  
        

```python
def frange(start,stop=None,step=0.1):
    list3=[]
    if(stop==None and step==0.1):
        stop=start
        start=0.0
    i=start
    while True:
        list3.append(float("%g" %i));
        i = i+step
        if(i>=stop):break
    return list3

print(frange(0.5))
print(frange(1.0,2.0))
print(frange(2.2,4.0,0.5))
```

#### [설명]  
- frange(start,stop,step) 함수에 매개변수 하나에만 값을 넣으면 함수 정의에 따르면 start가 아닌 stop에 들어가야 한다. 그래서 매개변수 1개만 입력됐을 때를 stop에 아무값도 안들어가서 stop값이 None이 되고 step값이 디폴트 값인 0.1인 경우로 생각하고 if문을 구현했다.매개변수가 1개이면 start에 값이 들어가지만 이 값이 stop에 들어가고 start는 기본값인 0.0이 되어야하므로 if문 내를 위와 같이 구현했다.  
- 매개변수가 1개만 입력된 경우 외에는 다 각자 매개변수 순서대로 적절히 입력되므로 다른 if문으로 다른 case를 처리하는 것은 만들지 않았다.  
- 변수 i에 start를 넣고 부동소수점에 의한 오차를 발생하지 않기 위해 float형을 짧게 만들어서 str으로 표현해주는 %g를 통한 포멧팅을 했고 포멧팅으로 str으로 변한 값을 다시 float로 형변환해 list3에 값을 추가했다.   
- i에 step값을 계속 더해서 위와 같은 과정을 반복했다.

          - 3  
      - sum(1,2,3,4,5)  
          - 15
      - sum(1,5,7,2,-10)

```python
def sum(*num):
    result=0
    for i in num:
        result+=i
    return result

print(sum())
print(sum(1,2))
print(sum(1,2,3,4,5))
print(sum(1,5,7,2,-10))
```

```python
def myinitial():
    s=input()
    A=s.split()
    X=list(range(len(A)))
    B=list(map(lambda x:A[x][0].upper(),X))
    print(''.join(B))
myinitial()
```

#### 문제 6  
 음이 아닌 정수n를 입력받으면 n!(factorial)을 계산하는 myfact()함수를 재귀적 함수로 구현하시오.  
   
   - [옵션] 참을 수 있는 정도 만큼의 수행시간을 직접 기다려보면서 n을 늘려보도록 합시다.  
   

```python
def myfact(n=-1):
    if n < 0 : 
        num=int(input())
        return myfact(num)
    elif n == 0 : return 1
    else:
        return n*myfact(n-1)
print(myfact())
```

#### [설명]  
- n<0 일 때를 설정해서 처음 n의 디폴트 값을 -1로 주므로써 num에 입력을 받게 했고  그 후에도 음수가 입력됐을 때도 이 조건문을 통해 다시 입력받을 수 있게 했다. 
- n!(factorial)은 n부터 1까지의 수들을 모두 곱한 값이므로 이를 재귀적으로 n*myfact(n-1)로 구현했다. 
- 함수가 종료되는 시점을 표기하기 위해 n==0일 때 1값을 반환하게 했다.

```python
import requests
import string
import os
import pickle

with open("E_stopwords.txt",'r') as f:
    S_E_stopwords=f.read()
with open("Kor_stopwords.txt",'r') as f:
    S_Kor_stopwords=f.read()
with open("sign.txt",'r') as f:
    S_sign_words=f.read()
E_stopwords=S_E_stopwords.split()
Kor_stopwords=S_Kor_stopwords.split('\n')
sign_words=S_sign_words.split()

urllist=['https://www.jetbrains.com','https://www.koreatech.ac.kr','http://www.naver.com'
         ,'http://www.daum.net','http://www.nytimes.com']
srclist=[]
for i in range(len(urllist)):
    srclist.append(requests.get(urllist[i]).text)

# 함수1 : JS,RSS,CSS 코드 제거 , Tag 제거
def text_parsing(src):
    i=0
    Pure_txt=''
    while i< len(src): 
        if src[i]=='<': # ①
            if src[i:i+7]=='<script': #①-②-1
                i+=7
                while src[i:i+9]!='</script>':
                    i+=1
                i+=9
            elif src[i:i+6]=='<style': #①-②-2
                i+=6
                while src[i:i+8]!='</style>':
                    i+=1
                i+=8
            else :   #①-②-3
                i+=1
                while src[i]!='>':
                    i+=1
                i+=1
        else:
            Pure_txt += src[i]
            i+=1
    return Pure_txt.split()
#함수2 : 순수 텍스트가 요소로 저장되어 있는 리스트에서 단어의 출현빈도를 사전으로 반환
def stopwords_filter(list):
    # 영어 불용어 필터
    for i in range(len(list)):
        for j in range(len(E_stopwords)):
            if list[i].lower()==E_stopwords[j]:
                list[i]=''
    # 한글 불용어 필터 
    ## 단어의 끝에서부터 불용어와 똑같은 길이만큼 슬라이싱해 불용어와 일치시 불용어 부분을 제거하여 저장
    for i in range(len(list)):
        for j in range(len(Kor_stopwords)):
            if len(list[i])>=len(Kor_stopwords[j]):
                if list[i][-len(Kor_stopwords[j]):]==Kor_stopwords[j]:
                    list[i]=list[i][:-len(Kor_stopwords[j])]

                
    # 특수문자 필터
    for i in range(len(list)):
        for j in range(len(sign_words)):
            list[i]=list[i].replace(sign_words[j],'')
    
    return list

#함수3 : 리스트 내 str 원소의 구두점 제거, 리스트 원소 값이 비어있으면 제거
def list_string_pure(list):
    for i in range(len(list)):
        for j in range(len(string.punctuation)):
            list[i]=list[i].replace(string.punctuation[j],'')
    
    list = stopwords_filter(list) # 불용어 제거
    
    i=0
    while i!=len(list):
        if list[i]=='':
            del list[i]
            continue
        i+=1
    return list

#함수4: 단어 count한 후 사전객체로 반환
def dict_countWords(list):
    d={}
    for i in range(len(list)):
        d[list[i]]=0
    for i in range(len(list)):
        d[list[i]]=d[list[i]]+1
    return d    

Dicts = [] # pickle로 load할 사전객체를 담을 빈 리스트 선언
dir = 'simpleSearchEngine_data' # 생성할 폴더 이름
if (not os.path.isdir(dir)): # 폴더 존재 유무 확인
    os.mkdir(dir)
for i in range(len(urllist)):
    # html 소스 동일 폴더에 저장
    ## 도메인에 '.html'붙여서 저장하기 위해 아래의 두가지 경우로 나눔
    if(urllist[i][:7]=='http://'):
        s = urllist[i][7:]+'.html'
    elif(urllist[i][:8]=='https://'):
        s = urllist[i][8:]+'.html'
    ### dir 폴더에 도메인명.html 파일을 입력
    with open(os.path.join(dir, s),'w', encoding='utf-8') as f:
        f.write(srclist[i])
     
    # 각 url의 단어 출현빈도 파일을 동일한 폴더에 생성
    P = list_string_pure(text_parsing(srclist[i])) # P는 구두점 제거, 비어있는 리스트 원소값 제거한 리스트
    D = dict_countWords(P) # 사전 객체 반환
    
    ## 도메인에 '.words_frequency.pickle'붙여서 저장하기 위해 
    ## 아래의 두가지 경우로 나눔
    if(urllist[i][:7]=='http://'):
        s = urllist[i][7:]+'.words_frequency.pickle'
    elif(urllist[i][:8]=='https://'):
        s = urllist[i][8:]+'.words_frequency.pickle'
    ### dir 폴더에 도메인명.words_frequency.pickle 파일을 입력
    with open(os.path.join(dir, s),'wb') as f:
        pickle.dump(D,f)
    with open(os.path.join(dir, s),'rb') as f:
        Dicts.append(pickle.load(f))
        
# 5개 사전객체 각각에 대해서 가장 많이 출현한 단어 3개씩 뽑기
print("[가장 많이 출현한 단어 3개 뽑기]")
for i in range(len(Dicts)):
    Result=[]
    list1 = list(Dicts[i].keys())
    list2 = list(Dicts[i].values())
    TupleD = tuple(zip(list1,list2))
    list2.sort(reverse=True) # 키값이 들어가 있는 list2를 역순으로 정렬
    
    # 정렬된 list2와 원래 사전 객체 순서로 저장된 튜플 TupleD[k][1]을 
    ## 비교했을 때 같으면 그 때의 key값을 넣으면 빈도수 가장 높은 
    ### 단어 순서로 Result 리스트에 추가된다.
    #### 따라서 많이 출현한 단어 3개를 뽑을 때 0부터 2까지의 Result 리스트 
    ##### 요소값을 뽑아내면 된다.
    for j in range(len(list2)):
        for k in range(len(TupleD)):
            if list2[j]==TupleD[k][1]: 
                Result.append(TupleD[k][0])
    print(urllist[i],':',Result[0]+','+Result[1]+','+Result[2])

def search():
    s = input("검색어를 입력하세요 : ")
    similarity=[0 for i in range(len(Dicts))]
    s_list = s.split()
    s_list = list_string_pure(s_list)
    # 사전 객체에 같은 단어가 있는 경우 단어의 count값을 similarity에 계속 더하고
    ## similarity 가 가장 높게 나온 사이트를 가장 먼저 출력한다.
    for i in range(len(Dicts)): 
        for j in s_list:
            for k in Dicts[i].keys():
                if j == k:
                    similarity[i] += Dicts[i][j]
    list=[]
    # 유사도가 똑같은 url이 두번 출력 안되게 한 번 출력된 url을 key값으로 저장할 사전객체
    ## 해당 url에 해당되는 value 가 1이 되면 출력이 안되고 다음 url을 확인하게 함
    tempDicts={}
    for i in range(len(urllist)):
        tempDicts[urllist[i]]=0
        
    for i in range(len(Dicts)):
        list.append((similarity[i],Dicts[i]))
    temp=[i for i in similarity]
    temp.sort(reverse=True)
    for i in range(len(temp)):
        for j in range(len(list)):
            if temp[i]==list[j][0]:
                if tempDicts[urllist[j]]==0:
                    print("유사도:",list[j][0],'->' ,urllist[j])
                    tempDicts[urllist[j]]=1
                    break
                elif tempDicts[urllist[j]]==1:
                    continue
search()
```
