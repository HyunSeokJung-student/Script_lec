### 문제 4
다음 문자열 ':'을 기준으로 분리하여 리스트로 만들고 각 문자열의 좌우 공백을 제거하시오  (즉, 문자열 S에서 I를 만들어라)

    s = '   first star   :   second star   :   third star  '
    I = ['first star', 'second star', 'third star']
    
- 1) for~in 문을 이용한 코드를 작성하시오.
- 2) 리스트 내포(list comprehension)을 이용한 코드를 작성하시오.

```python
#4-1
s4='   first star   :   second star   :   third star   '
I4=s4.split(':')
for i in range(len(I4)):
    I4[i]=I4[i].strip()
print(I4)
```

[설명]  

    split(':')을 통해 단어와 공백이 하나의 원소로 이뤄져 있는 I4리스트를 만들었다. 이 때 I4는 ['   first star   ', '   second star   ', '   third star   ']가 된다. 이 때 문자열 메소드 strip()을 사용하면 좌우공백이 지워진다. 그렇기 때문에 리스트의 각 원소 값에 strip()을 써서 다시 리스트에 저장하면 ['first star', 'second star', 'third star']을 얻을 수 있다.  
    
    이 때 for~in을 사용해서 하려면 i를 리스트 I4의 인덱스라고 했을 때 i를 0부터 len(I4)-1까지 range(len(I4))로 설정해서 i를 통해 리스트 각각의 원소에 접근해서 strip()을 쓰고 그 결과로 반환된 문자열을 I4[i]에 바로 저장하는 방식을 쓸 수 있다.

```python
#4-2
s4='   first star   :   second star   :   third star   '
I4=s4.split(':')
I4=[ I4[i].strip() for i in range(len(I4))]
print(I4)
```

[설명]  

    split(':')을 통해 단어와 공백이 하나의 원소로 이뤄져 있는 I4리스트를 만들었다. 이 때 I4는 ['   first star   ', '   second star   ', '   third star   ']가 된다. 이 때 문자열 메소드 strip()을 사용하면 좌우공백이 지워진다. 그렇기 때문에 리스트의 각 원소 값에 strip()을 써서 다시 리스트에 저장하면 ['first star', 'second star', 'third star']을 얻을 수 있다.  
    
    이 때 list comprehension을 사용하려면 위와 마찬가지로 i로 리스트 I4의 인덱스를 설정하고 각 리스트의 원소 값에 strip()을 적용한 문자열을 리스트로 구성한 후 구성된 리스트 전체를 리스트 I4에 저장하는 방식을 쓸 수 있다.

### 문제 5
다음과 같이 0보다 큰 정수 리스트 변수 list를 인자로 받는 함수 addall(list)와 addallodd(list)를 for ~ in 리터럴과 리스트 내포 방식으로 각각 제시하시오.
- 1) addall(list) 함수 (리스트 내의 모든 정수를 더하는 함수로서 해답으로는 for~in  리터럴과 리스트 내포 방식으로 각각 제시하시오.)
    -  
    
            >>> addall([1])  
            1
    -  
    
            >>> addall([1,2,3,4,5,6,7,8,9])
            45
          
- 2) addallodd(list)함수 (리스트내의 모든 홀수를 더하는 함수로서 해답으로는 for~in리터럴과 리스트 내포 방식으로 각각 제시하시오.)
    -  
    
            >>> addallodd([1])  
            1
    -  
    
            >>> addallodd([1,2,3,4,5,6,7,8,9])
            25
            
- [참고 1]: 리스트 내포 방식으로 addall() 및 addallodd()함수를 만들 때엔 리스트 내포에 의해 생성된 리스트 결과에 대해 내장 함수를 사용하는 방식을 고려해보시오 (주의: 리스트 내포 방식의 addall() 및 addallodd() 함수의 몸체는 단 1줄로 작성되어야 한다.)
- [참고 2]: sum() 등의 내장함수 사용 가능함

```python
#5-1-1 for~in
def addall(list) :
    res = 0
    for i in range(len(list)) :
        res += list[i]
    return res
print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))
```

[설명]  

    함수를 만들 때 결과를 저장하는 res를 0으로 초기화한 후 for~in 을 통해 i로 리스트의 인덱스 요소값에 접근해서 for~in 루프가 돌 때마다 list[i]값을 res에 더하게 하고 이를 반환하게 하여 함수 addall(list)를 만들 수 있었다.

```python
#5-1-2 list comprehension
def addall(list) :
    return sum([list[i] for i in range(len(list))])
print(addall([1]))
print(addall([1,2,3,4,5,6,7,8,9]))
```

[설명]  

    리스트 내포를 이용해 리스트의 모든 원소 값을 원소로 갖는 리스트를 생성한 후 이를 리스트 내장 함수 sum()을 이용해 모든 원소의 합을 구해서 이를 반환하므로써 addall(list)를 구현할 수 있었다.

```python
#5-2-1 for~in
def addallodd(list):
    res =0
    for i in range(len(list)):
        if list[i]%2 == 1:
            res += list[i]
    return res
print(addallodd([1]))
print(addallodd([1,2,3,4,5,6,7,8,9]))
```

[설명]  

    for~in 에서 변수 i로 리스트의 모든 원소 값에 접근할 수 있게 range()를 설정해주고 홀수인 경우만 res(결과)에 더해주기 위해 if문에 홀수가 되는 조건(list[i]%2==1)을 설정해서 res를 반환하는 addallodd(list)함수를 구현할 수 있었다.

```python
#5-2-2 list comprehension
def addallodd(list):
    return sum([list[i] for i in range(len(list)) if list[i]%2==1 ])
print(addallodd([1]))
print(addallodd([1,2,3,4,5,6,7,8,9]))
```

[설명]  

    리스트 내포에서 i를 통해 리스트의 모든 원소 값에 접근할 수 있게 (for i in range(len(list))로 설정했고 이 때 리스트 내포를 통해 새로 생성되는 원소에 들어가는 값을 홀수만 들어가게 하기 위해 리스트 내포 내에 (if list[i]%2==1)을 적었다. 그 후 리스트 내포로 홀수만을 원소로 가지는 리스트를 sum()함수를 이용해 리스트 전체의 합을 구해서 반환하는 addallodd(list)함수를 구현할 수 있었다.

### 문제 7
사전 d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}가 주어졌을 때 다음 요구사항에 맞는 코드를 제시하시오.  
- [참고]: d.keys(), d.values()를 통해 리스트를 얻어낸 후 리스트가 지니고 있는 sort(cmp=None, key=None, reverse=False)함수를 활용하시오.  
- 1) 키의 알파벳 오름차순 순서대로 튜플 (키,값)을 차례대로 출력하시오.  
- 2) 키의 알파벳 내림차순 순서대로 튜플 (키,값)을 차례대로 출력하시오.  
- 3) 값의 오름차순 순서대로 튜플 (키,값)을 차례대로 출력하시오.  
- 4) 값의 내림차순 순서대로 튜플 (키,값)을 차례대로 출력하시오.

```python
#7-1 키 알파벳 오름차순
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
keys = list(d.keys())
keys.sort()
T1=()
for i in keys:
    T1=T1+((i,d[i]),)
print(T1)
```

[설명]  
   
     d.keys()로 나온 d.keys 객체를 list로 변환 후 리스트 keys에 저장했다. 리스트 내장함수 sort()로 해당 리스트 내 원소를 오름차순으로 정렬하게 하고 리스트 내 원소를 하나씩 사전 d의 키와 일치하는 값을 짝지어 튜플 형태로 다시 저장되게 했다.

```python
#7-2 키 알파벳 내림차순
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
keys = list(d.keys())
keys.sort(reverse=True)
T2=()
for i in keys:
    T2=T2+((i,d[i]),)
print(T2)
```

[설명]  
        
     d.keys()로 나온 d.keys 객체를 list로 변환 후 리스트 keys에 저장했다. 리스트 내장함수 sort(reverse=True)로 해당 리스트 내 원소를 내림차순으로 정렬하게 하고 리스트 내 원소를 하나씩 사전 d의 키와 일치하는 값을 짝지어 튜플 형태로 다시 저장되게 했다.

```python
#7-3 값 오름차순
d = {'one':1, 'two':2, 'three':3, 'four':4, 'five':5}
values = list(d.values())
keys = list(d.keys())
values.sort()
T3=()
for i in values:
    for j in keys:
        if i==d[j]:
            T3=T3+((j,d[j]),)
print(T3)
```

[설명]  

    d.keys()로 나온 d.keys 객체를 list로 변환 후 리스트 keys에 저장했다. 또 d.values()로 나온 d.values 객체를 list로 변환 후 리스트 values에 저장했다. 그 후, 리스트 value를 sort()함수를 통해 값을 오름차순으로 정렬하고 그렇게 정렬된 값을 리스트 keys에 저장된 모든 키와 비교하는 for문을 사용해서 키와 값이 일치하는 경우 이를 튜플로 짝지어 튜플 형태로 T3에 저장되게 했고 이를 출력했다.

### 문제 8 [ACM-ICPC 1673]  
- 치킨 쿠폰  
  
      강민이는 치킨 한 마리를 주문할 수 있는 치킨 쿠폰을 n장 가지고 있다. 이 치킨집에서는 치킨을 한 마리 주문할 때마다 도장을 하나씩 찍어 주는데, 도장을 k개 모으면 치킨 쿠폰 한 장으로 교환할 수 있다.

        강민이가 지금 갖고 있는 치킨 쿠폰으로 치킨을 최대 몇 마리나 먹을 수 있는지 구하여라. 단, 치킨을 주문하기 위해서는 반드시 치킨 쿠폰을 갖고 있어야 한다.  
     
![image](https://github.com/JungHyunSuok/Script_lec/blob/master/ACMICPC1673.png?raw=true)

```python
while True:
    try:
        p8_N,p8_K=map(int,input().split())
        p8_N=int(p8_N)
        p8_K=int(p8_K)
        temp=p8_N
        res=p8_N
            
        while True:
            temp1=(temp//p8_K)
            temp2=(temp%p8_K)
            temp=temp1+temp2
            res+=temp1
            if temp1<1:
                break
        print(res)
    except: exit()
```

[설명]  
  
    처음 치킨을 시킬 때는 N마리를 시키고 N개의 도장을 받을 수 있다. 그 후 N개의 도장으로 시킬 수 있는 치킨 수는 N//K(temp1)이고 사용하지 못한 도장의 수는 N%K(temp2) 이다. 또 후에 치킨을 시킬 때 전에 사용하지 못한 도장N%K(temp2)와 N//K(temp1)로 받은 도장을 합한 temp를 K로 나누면 시키기 전 도장 수로 시킬 수 있는 치킨 수가 나온다. 이 과정을 temp1이 1보다 작을 때까지 반복하면 지금 가진 도장 수로 시킬 수 있는 최대 치킨 수를 구할 수 있다.

  - 요구사항 1.CSS와 Javascript 내용도 모두 제거하시오   
      
          º 즉, 아래와 같은 코드는 <style> 및 <script> 태그 뿐만 아니라 그 안의 내용 전체를 모두 제거

        º 예:['world',':','+hello+','~python$$$','=','world'] ---> ['world','hello', 'python', 'world']   
        º 모든 punctuation을 확인하는 방법은 아래와 같다.  
                import string

- 요구사항 3. 만약 punctuation (구두문자)를 제거한 이후 공백으로 판명되는 단어들은 해당 단어 리스트에서 제거하시오.  
  
         º 예:['컴퓨터공학부!','---?', '%지향~', '...'] ---> ['컴퓨터공학부', '', '지향', ''] ---> ['컴퓨터공학부', '지향']   
         
- 요구사항 3. 각 단어들의 출현빈도를 사전형태로 저장하여 출력하시오.
  
         º 예: ['world', 'hello', 'python', 'world', '컴퓨터공학부', '지향'] ---> {'world': 2, 'hello': 1, 'python': 1, '컴퓨터공학부': 1, '지향': 1}   
          
- 요구사항 4. 다음 5개의 사이트에 대한 단어 출현빈도 결과를 출력하시오.  
º http://cse.koreatech.ac.kr  
º https://www.koreatech.ac.kr  
º http://www.naver.com  
º http://www.daum.net  
º http://www.nytimes.com 

```python
import requests
import string

req1 = requests.get('http://cse.koreatech.ac.kr')
source1 = req1.text 
req2 = requests.get('https://www.koreatech.ac.kr')
source2 = req2.text
req3 = requests.get('http://www.naver.com')
source3 = req3.text
req4 = requests.get('http://www.daum.net')
source4 = req4.text
req5 = requests.get('http://www.nytimes.com')
source5 = req5.text

# 함수1 : JS,RSS,CSS 코드 제거 , Tag 제거
def text_parsing(src):
    i=0
    Pure_txt=''
    while i< len(src): 
        if src[i]=='<': # ①
            if src[i:i+7]=='<script': #①-②-1
                i+=7
                while src[i:i+9]!='</script>':
                    i+=1
                i+=9
            elif src[i:i+6]=='<style': #①-②-2
                i+=6
                while src[i:i+8]!='</style>':
                    i+=1
                i+=8
            else :   #①-②-3
                i+=1
                while src[i]!='>':
                    i+=1
                i+=1
        else:
            Pure_txt += src[i]
            i+=1
    return Pure_txt.split()

#함수2 : 리스트 내 str 원소의 구두점 제거, 리스트 원소 값이 비어있으면 제거
def list_string_pure(list):
    for i in range(len(list)):
        for j in range(len(string.punctuation)):
            list[i]=list[i].replace(string.punctuation[j],'')
    i=0
    while i!=len(list):
        if list[i]=='':
            del list[i]
        i+=1
    return list
#함수3 : 순수 텍스트가 요소로 저장되어 있는 리스트에서 단어의 출현빈도를 사전으로 반환
def dict_countWords(list):
    d={}
    for i in range(len(list)):
        d[list[i]]=0
    for i in range(len(list)):
        d[list[i]]=d[list[i]]+1
    return d    
print("http://cse.koreatech.ac.kr의 결과")
P1=list_string_pure(text_parsing(source1)) # P는 구두점 제거, 비어있는 리스트 원소값 제거한 리스트
print(dict_countWords(P1))
print()
print("https://www.koreatech.ac.kr의 결과")
P2=list_string_pure(text_parsing(source2))
print(dict_countWords(P2))
print()
print("http://www.naver.com의 결과")
P3=list_string_pure(text_parsing(source3))
print(dict_countWords(P3))
print()
print("http://www.daum.net의 결과")
P4=list_string_pure(text_parsing(source4))
print(dict_countWords(P4))
print()
print("http://www.nytimes.com의 결과")
P5=list_string_pure(text_parsing(source5))
print(dict_countWords(P5))
```

[설명]  

    # 함수1

    코드와 태그를 제거하기 위해 '<'로 시작해서 '>'을 만날 때까지의 char를 다 무시하는 경우와 '<'로 시작해서 '</ >'까지의 char를 다 무시하는 경우로 나눌 수 있다.   
    
    이를 더 구체적으로 나누면 다음과 같이 나눌 수 있다.
        ①:str[i]이'<'일 때와 아닐 때

        ①-②-2:str[i:i+6]이 '<style' 일 때
